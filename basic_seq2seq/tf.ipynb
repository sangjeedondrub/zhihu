```python
import tensorflow as tf
```

```python
x = tf.Variable(tf.random_normal([5,3]),name='x')
y = tf.Variable(tf.random_normal([3,5]),name='y')
mul = tf.matmul(x,y)
```

```python
init_op = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init_op)
    res = sess.run(mul)
    
    print(res)
    
    restile = tf.tile(res,[5,3])
    print(sess.run(restile))
```
